**Часть 1.** Напишите сначала функцию генерации случайных величин из равномерного распределения на отрезке $[0, 1]$ с заданной точностью. Это можно сделать, записав случайную величину $\xi \sim U[0, 1]$ в двоичной системе системе счисления $\xi = 0,\xi_1\xi_2\xi_3...$. Тогда $\xi_i \sim Bern(1/2)$ и независимы в совокупности. Приближение заключается в том, что вместо генерации бесконечного количества $\xi_i$ мы полагаем $\xi = 0,\xi_1\xi_2\xi_3...\xi_n$.

Нужно реализовать функцию нужно так, чтобы она могла принимать на вход в качестве параметра `size` как число, так и объект `tuple` любой размерности, и возвращать объект `numpy.array` соответствующей размерности. Например, если `size=(10, 1, 5)`, то функция должна вернуть объект размера $10 \times 1 \times 5$. Кроме того, функцию `coin` можно вызвать только один раз, и, конечно же, не использовать какие-либо циклы. Аргумент `precision` отвечает за число $n$.

```python
def uniform(size=1, precision=30):
    return ((coin(np.prod(size) * precision)).reshape(np.prod(size), precision)).dot(2 ** np.arange(0, precision)[::-1]) / (2 ** precision - 1)

uniform((5, 5), 50)
```

Для $U[0, 1]$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-0.25, 1.25]$, а также гистограмму по сгенерированным случайным величинам.

```python
size = 200
grid = np.linspace(-0.25, 1.25, 500)

# sample = sps.uniform(loc=0, scale=1).rvs(size)
sample = uniform(size, precision=50)

# Отрисовка графика
plt.figure(figsize=(10, 4))

# отображаем значения случайных величин полупрозрачными точками
plt.scatter(
    sample, 
    np.zeros(size), 
    alpha=0.4, 
    label='Значения'
)

# по точкам строим нормированную полупрозрачную гистограмму
plt.hist(
    sample, 
    bins=10, 
    density=True,
    alpha=0.4, 
    color='orange', 
    label='Гистограмма'
)

# рисуем график плотности
plt.plot(
    grid, 
    sps.uniform.pdf(grid), 
    color='red', 
    lw=3, 
    label='Плотность'
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

**Часть 2.** Напишите функцию генерации случайных величин в количестве `size` штук из распределения $\mathcal{N}(loc, scale^2)$ с помощью преобразования Бокса-Мюллера, которое заключается в следующем. Пусть $\xi$ и $\eta$ — независимые случайные величины, равномерно распределенные на $(0, 1]$. Тогда случайные величины $X = cos(2\pi\xi)\sqrt{-2\ln\eta}, Y = sin(2\pi\xi)\sqrt{-2\ln\eta}$ являются независимыми нормальными $\mathcal{N}(0, 1)$.

Реализация должна быть без циклов. В качестве параметра `size` может быть как число, так и объект `tuple` любой размерности. Желательно использовать как можно меньше бросков монеты. Можно использовать реализованную ранее функцию `uniform`.

```python
settings = np.seterr(all='ignore')

def fun_x(pair):
    return np.cos(2 * np.pi*pair[0]) * np.sqrt(-2 * np.log(pair[1]))

def fun_y(pair):
    return np.sin(2 * np.pi*pair[0]) * np.sqrt(-2 * np.log(pair[1]))

def fun_x_rev(pair):
    return np.cos(2 * np.pi*pair[1]) * np.sqrt(-2 * np.log(pair[0]))

def fun_y_rev(pair):
    return np.sin(2 * np.pi*pair[1]) * np.sqrt(-2 * np.log(pair[0]))
    
def normal(size=4, loc=0, scale=1, precision=30):
    #утверждается, что достаточно 2 массивов значений по sqrt(size / 2)
    #тогда рассотрев каждое значение ksi с каждым eta, получим size / 2 * 2 значений для номрального распределения     
    #можно было бы пойти дальше и обойтись sqrt(size) / 2 значениями, если ksi и eta поменять местами в формулах    
    #во избежании проблем с округлениями будем считать больше и обрезать     
    ksi = uniform(int(np.sqrt(np.prod(size) / 2)), precision)
    eta = uniform(int(np.sqrt(np.prod(size)) / 2), precision)
    #попарно 
    all_pairs = np.array(np.meshgrid(ksi, eta)).T.reshape(-1,2)
    
    vec1 = np.apply_along_axis(fun_x, 1, all_pairs)
    vec2 = np.apply_along_axis(fun_y, 1, all_pairs)
    vec3 = np.apply_along_axis(fun_x_rev, 1, all_pairs)
    vec4 = np.apply_along_axis(fun_y_rev, 1, all_pairs)
    
    vec = np.concatenate((vec1, vec2, vec3, vec4))
    
    vec = vec * (scale ** 2) + loc
    #оберзаем лишние     
    vec = vec[0:size]
    
    assert(len(vec) == np.prod(size))
    
    return vec
    
normal(size=1000)
```

Для $\mathcal{N}(0, 1)$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-3, 3]$, а также гистограмму по сгенерированным случайным величинам.

```python
size = 200
grid = np.linspace(-3, 3, 500)
# sample = sps.norm(0, 1).rvs(size)
sample = normal(size, loc=0, scale=1)
sample = normal(size, loc=0, scale=1)

# Отрисовка графика
plt.figure(figsize=(10, 4))

# отображаем значения случайных величин полупрозрачными точками
plt.scatter(
    sample, 
    np.zeros(size), 
    alpha=0.4, 
    label='Значения'
)

# по точкам строим нормированную полупрозрачную гистограмму
plt.hist(
    sample, 
    bins=20, 
    density=True,
    alpha=0.4, 
    color='orange',
    label='Гистограмма'
)

# рисуем график плотности
plt.plot(
    grid, 
    sps.norm.pdf(grid), 
    color='red', 
    lw=3, 
    label='Плотность'
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

На мой взгляд, такой выборки недостаточно, слишком разнообразными получаются графики при разных запусках, поэтому предлагаю посомтреть на график, получаемый при $size=20000$

```python
size = 20000
grid = np.linspace(-3, 3, 500)
# sample = sps.norm(0, 1).rvs(size)
sample = normal(size, loc=0, scale=1)

# Отрисовка графика
plt.figure(figsize=(10, 4))

# отображаем значения случайных величин полупрозрачными точками
plt.scatter(
    sample, 
    np.zeros(size), 
    alpha=0.4, 
    label='Значения'
)

# по точкам строим нормированную полупрозрачную гистограмму
plt.hist(
    sample, 
    bins=20, 
    density=True,
    alpha=0.4, 
    color='orange'
)

# рисуем график плотности
plt.plot(
    grid, 
    sps.norm.pdf(grid), 
    color='red', 
    lw=3, 
    label='Плотность'
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

**Часть 3.** Вы уже научились генерировать выборку из равномерного распределения. Напишите функцию генерации выборки из экспоненциального распределения, используя задачу из теории вероятностей:

*Если $\xi$ — случайная величина, имеющая абсолютно непрерывное распределение, и $F$ — ее функция распределения, то случайная величина $F(\xi)$ имеет равномерное распределение на $[0, 1]$*.

Какое преобразование над равномерной случайной величиной необходимо совершить? Поясните ответ.

**Необходим метод обратного преобразования.**

Пусть $U ∼ {U}[0, 1]$. Рассмотрим случайную величину $\xi$, такую что $\xi_i = \{x : F(x) = U_i\}$. $F(x)$ - строго монотонная функция распределения экспоненциального распредления $F(x) = 1 - e^{-\lambda x}$, поэтому $\xi_i$ - это всегда одна точка.
$\mathbb{P}\{\xi_i \leq x\} = \mathbb{P} \{F^{-1}(U_i) \leq x\} = \mathbb{P} \{U_i \leq F(x)\} = F_U(F(x)) = F(x)$, так как $F_u(y) = y$. То есть $\xi$ имеет функцию распределения $F$, оно задаёт однозначно распредлеение. Значит $\xi$ будет иметь экспоненциальное распредление и будет вычисляться как обратная к $F$, $\xi_i = F^{-1}(U_i) = -\frac{1}{\lambda}\ln(1-U_i)$

Для получения полного балла реализация должна быть без циклов, а параметр `size` может быть типа `tuple`.

```python

def expon(size=1, lambd=1, precision=30):
    u = uniform(size, precision)
    return -1 / lambd * np.log(1 - u)
```

Для $Exp(1)$ сгенерируйте выборку размера 100 и постройте график плотности этого распределения на отрезке $[-0.5, 5]$.

```python
size = 100
grid = np.linspace(-0.5, 5, 500)
# sample = sps.exp(1).rvs(size)
sample = expon(size, lambd=1)

# Отрисовка графика
plt.figure(figsize=(10, 4))

# отображаем значения случайных величин полупрозрачными точками
plt.scatter(
    sample, 
    np.zeros(size), 
    alpha=0.4, 
    label='Значения'
)

# по точкам строим нормированную полупрозрачную гистограмму
plt.hist(
    sample, 
    bins=30, 
    density=True,
    alpha=0.4, 
    color='orange',
    label='Гистограмма'
)

# рисуем график плотности
plt.plot(
    grid, 
    sps.expon.pdf(grid), 
    color='red', 
    lw=3, 
    label='Плотность'
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

Уже при значениях $size=100$ получаем выборку и гистограмму, которая согласуется с теоретической плотностью. При увеличении $size$ или $precision$ гистограмма будет улучшаться. ### <font color="orange"><i>Сложная часть: броуновское движение</i></font>

### Задача 3

<a target="_blank" href="https://ru.wikipedia.org/wiki/Броуновское_движение">Броуновское движение</a> — беспорядочное движение микроскопических видимых взвешенных частиц твёрдого вещества в жидкости или газе, вызываемое тепловым движением частиц жидкости или газа. 

Математически броуновское движение обычно моделируется как некоторая случайная функция $X_t$ аргумента $t$, который несет смысл времени. В каждый фиксированный момент времени $t$, значение $X_t$ является случайной величиной, характеризующей положение частицы в момент времени $t$. Считается, что для любого интервала времени $(s, t)$ перемещение (точнее, приращение) частицы за этот интервал времени имеет нормальное распределение: $X_t - X_s \sim \mathcal{N}(0, (t-s)\sigma^2)$. Причем величина $X_t - X_s$ не зависит от приращения частицы за другие интервалы времени, не пересекающиеся с интервалом $(s, t)$. Математическая корректность данного определения будет доказана в курсе случайных процессов.

Многомерное броуновское движение соответствует одномерным по каждой координате.

Сгенерировать траекторию частицы в интервале времени $(0, T)$ в предположении броуновского движения можно следующим образом:
1. Выбрать начальное положение, обычно это начало координат.
2. Разбить интервал $(0, T)$ равномерно на интервалы длины $\Delta t$. Величина $\Delta t$ должна быть достаточно мала, в частности, на графиках не должно быть видно дискретности компьютерной генерации. 
3. Сгенерировать приращение частицы на каждом интервале по каждой координате независимо из распределения $\mathcal{N}(0, \sigma^2 \Delta t)$.
4. Для получения значений $X_t$ просуммировать необходимые приращения.



**1.**
Реализуйте функцию семплирования траекторий броуновского движения. Функция должна принимать в качестве аргументов:
- Параметр $\sigma$;
- Время $T$;
- Количество интервалов длины $\Delta t$;
- Размерность пространства, в котором реализуется траектория;
- Количество различных частиц, для которых нужно сгенерировать траектории.

Функция должна вернуть трехмерный массив `result`, где `result[i, j, k]` &mdash; значение $j$-й координаты $i$-й частицы на $k$-м шаге.

При реализации считайте, что все частицы в начальный момент времени находятся в начале координат.
Функцию нужно реализовать для произвольной размерности, не используя циклы. Можно использовать функции из `scipy`.

```python

def brounian(sigma=1, time=100, num_intervals=100, dimension = 2, num_particles=1):
    delta_time = time / num_intervals     
    #возможность проверить на написанном нормальном распределении     
    #deltas = np.array(normal(size=num_particles * dimension * num_intervals, loc=0, scale=np.sqrt(delta_time) * sigma,  precision=60))
    deltas = np.array(sps.norm(loc=0, scale=np.sqrt(delta_time) * sigma).rvs(size=num_particles * dimension * num_intervals))
    deltas = deltas.reshape(num_particles, dimension, num_intervals)
    deltas = deltas.cumsum(axis=2)
    return deltas
    
brounian(1, 100, 15, 2, 1)

```

**3.** Исследуйте, как зависит среднее расстояние частицы до начала координат от времени и от параметра $\sigma$. Проведите исследование для разных размерностей пространства. Под средним в теории понимается математическое ожидание. На практике нужно исследовать его оценку посредством генерации множества траекторий.

В исследовании обязательно дожны присутствовать текстовые пояснения и графики. Сделайте выводы.

Зависимость среднее расстояние частицы до начала координат от времени.

```python
def distance_from_time(dim=2, num_of_particles=100):
    times = np.arange(0, 1000, step=10)
    k = 1000
    result = np.zeros(shape=(1, len(times)))
    result = np.insert(result, 1, np.array(times), axis=0)
    
    for i, time in enumerate(times):
        dist_for_cur_time = 0
        sample = brounian(1, time, num_intervals=k, dimension=dim, num_particles=num_of_particles)
        
        for j in range(num_of_particles):
            dist_for_cur_time += np.sqrt(np.sum(sample[j,0:dim,k-1]**2))
        
        dist_for_cur_time /= num_of_particles
        
        result[0, i] = dist_for_cur_time
    
    return result
```

```python
sample1 = distance_from_time(dim = 1) 
sample2 = distance_from_time(dim = 2) 
sample3 = distance_from_time(dim = 3) 
```

```python
sqrt_x = np.linspace(0, 1000, 100)
sqrt_y1 = np.sqrt(sqrt_x) * 0.8

# Отрисовка графика
plt.figure(figsize=(12, 6))

# отображаем значения случайных величин полупрозрачными точками
plt.scatter(
    sample1[1],
    sample1[0], 
    alpha=0.4, 
    label='1 - мерное',
    color='blue'
)

plt.scatter(
    sample2[1],
    sample2[0], 
    alpha=0.4, 
    label='2- мерное',
    color='red'
)

plt.scatter(
    sample3[1],
    sample3[0], 
    alpha=0.4, 
    label='3 - мерное',
    color='green'
)


plt.xlabel('time')
plt.ylabel('dx')

plt.plot(
    sqrt_x, 
    sqrt_y1,
    label='y= x^(1/2) * c',
    color='blue'
)

plt.legend()
plt.grid(ls=':')
plt.show()
```

```python
 def distance_from_sigma(dim=2, num_of_particles=100):
    sigmas = np.arange(0, 5, step=0.1)
    k = 1000
    result = np.zeros(shape=(1, len(sigmas)))
    result = np.insert(result, 1, np.array(sigmas), axis=0)
    
    for i, sigma in enumerate(sigmas):
        dist_for_cur_time = 0
        
        sample = brounian(sigma=sigma, time=10, num_intervals=k, dimension=dim, num_particles=num_of_particles)
        
        for j in range(num_of_particles):
            dist_for_cur_time += np.sqrt(np.sum(sample[j,0:dim,k-1]**2))
        
        dist_for_cur_time /= num_of_particles
        
        result[0, i] = dist_for_cur_time 
    return result
```

```python
sample1 = distance_from_sigma(dim = 1) 
sample2 = distance_from_sigma(dim = 2) 
sample3 = distance_from_sigma(dim = 3) 
```

В ходе этой части мы явно получили зависимость смещения от времени $t$  и от $\sigma$ при разных размерностях пространства.
Зависимость от времени имеет вид $\sqrt{\langle x^{2} \rangle} \sim \sqrt{t}$,
Зависимость от $\sigma$ : $\sqrt{\langle x^{2} \rangle} \sim {\sigma}$.
Данный результат отлично согласуется с классической теорией Эйнштейна, по которой
$\sqrt{\langle x^{2} \rangle} = 2 \cdot D \tau$, где $D = \frac{RT}{6N_{A}\pi a \xi}$. В свою очередь классическая теория Эйнштейна согласуется с результатми физиков, полученным на практике.
Также стоит отметить, что, при увеличении рамзмерности пространства, смещение также увеличивается. Этого следовало ожидать, так как по каждой из осей в каждый интервал времени cмещение - независимая величина.

Распределения:

Для выполнения данного задания можно использовать код с лекции. Можно использовать функции из `scipy`. Сделать виджеты недостаточно, поскольку они не сохраняются.

```python
# Распределение бернулли
fig, axs = plt.subplots(ncols=4, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle('Плотность распределения Бернулли')

probs = [0.1, 0.3, 0.5, 0.7]

for i, p in enumerate(probs):
    x = [0, 1]
    y = [1-p, p]
    custm = sps.rv_discrete(name='custm', values=(x, y))
    
    axs[i].plot(
        x, 
        [1-p, p], 
        'ro', 
        ms=12, 
        mec='r'
    )
    
    axs[i].vlines(x, 0, custm.pmf(x), colors='r', lw=4)
    axs[i].set_title("p - {}".format(p))
```

```python
# выборка
size = 100
fig, axs = plt.subplots(ncols=4, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle('Выборка')

probs = [0.1, 0.3, 0.5, 0.7]

for i, p in enumerate(probs):
    sample = sps.bernoulli(p).rvs(size)

    axs[i].hist(
    sample, 
    bins=20, 
    alpha=0.4, 
    color='orange'
    )

    axs[i].set_title("p - {}".format(p))
```

Из курса теории вероятностей известно, что распределение Бернулли $\xi \sim Bern(p)$ моделирует случайный эксперимент при заранее известной вероятности успеха или неудачи. С вероятностью $p$ происходит успех, иначе - неудача с вероятностью $1- p$. Чем больше $p$, тем больше будет левый столбик гистограммы, так как веротяность успеха выше. Полученные зависимости полностью согласуются с теоретическими знаниями.

```python
n=20
fig, axs = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Плотность биноминального распредления при разных p')

probs = [0.25, 0.5, 0.75]

for i, p in enumerate(probs):
    x = np.arange(sps.binom.ppf(0.01, n, p),
              sps.binom.ppf(0.99, n, p))
    axs[i].plot(x, sps.binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
    axs[i].vlines(x, 0, sps.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
    
    axs[i].set_title("p - {}".format(p))
```

```python
p=0.5
fig, axs = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Плотность биноминальное распредления при разных n')

nums = [5, 10, 50]

for i, n in enumerate(nums):
    x = np.arange(sps.binom.ppf(0.01, n, p),
              sps.binom.ppf(0.99, n, p))
    axs[i].plot(x, sps.binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
    axs[i].vlines(x, 0, sps.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
    
    axs[i].set_title("n - {}".format(n))
```

Из курса теории вероятностей известно, что биноминальное распределение — распределение количества «успехов» в последовательности из $n$ независимых случайных экспериментов, таких, что вероятность «успеха» в каждом из них постоянна и равна $p$. Отсюда уже интуитивно понятно понтяно, что максимальная плотность должна быть в точке $p \cdot n$, поэтому при увеличении каждого из параметров график будет свдигаться вправо. При увеличении $n$ расширяется(график) и количество возможных значений случайной величины, а значит столбики должны уменьшится, что в сумме их длины давали единицу. А при увеличении $p$, он, наоборот, сужается. Все это отлично отражает формулу $\xi \sim Bin(n, p), P(\xi = k) = \binom{n}{k}\cdot p^{k} (1-p)^{n-k}$ 

```python
# выборка
size = 1000
n=10
fig, axs = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle('Выборка')

probs = [0.1, 0.2, 0.5]

for i, p in enumerate(probs):
    sample = sps.binom(n, p).rvs(size)
    axs[i].hist(
    sample, 
    bins=10, 
    alpha=0.4, 
    color='orange'
    )

    axs[i].set_title("p - {}".format(p))
```

```python
# Геометрическое распредление
fig, axs = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
fig.suptitle('Плотность геометрического распредления')

probs = [0.2, 0.3, 0.5]

for i, p in enumerate(probs):
    
    x = np.arange(sps.geom.ppf(0.01, p),
              sps.geom.ppf(0.99, p))
    axs[i].plot(x, sps.geom.pmf(x, p), 'bo', ms=8, label='geom pmf')
    axs[i].vlines(x, 0, sps.geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
    axs[i].set_title("p - {}".format(p))

    
```

```python
# выборка
size = 100
n=5
fig, axs = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle('Выборка')

probs = [0.1, 0.3, 0.5]

for i, p in enumerate(probs):
    sample = sps.geom(p).rvs(size)
    axs[i].hist(
    sample, 
    bins=30, 
    alpha=0.4, 
    color='orange'
    )

    axs[i].set_title("p - {}".format(p))
```

Из курса теории вероятностей известно, что геометрическое распределение — распределение вероятностей случайной величины $\xi$ равной номеру первого «успеха» в серии испытаний Бернулли. Отсюда уже интуитивно понятно, что максимальная плотность должна быть в точке $1$. При увеличении $p$ первые столбики будт расти, так как увеличивается вероятность успеха. Дальние должны уменьшаться, чтобы в сумме опять была единицы. Все это отлично отражает формулу $\xi \sim Geom(p), P(\xi = k) = (1-p)^{k-1} \cdot p$ 

```python
fig, axs = plt.subplots(ncols=1, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(12)
fig.suptitle('Плотность дискретного равномерного распределения')
low, high = 1, 10


x = np.arange(sps.randint.ppf(0.01, low, high),
          sps.randint.ppf(0.99, low, high))
axs.plot(x, sps.randint.pmf(x, low, high), 'bo', ms=8, label='randint pmf')
axs.vlines(x, 0, sps.randint.pmf(x, low, high), colors='b', lw=5, alpha=0.5)
```

```python
# выборка
size = 2000
fig, axs = plt.subplots(ncols=1, nrows=1, sharex=True, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(10)
fig.suptitle('Выборка')


sample = sps.randint(low, high).rvs(size)
axs.hist(
sample, 
bins=high-low, 
alpha=0.4, 
color='orange'
)

 
```

Из курса теории вероятностей известно, что дискретное равномерное распределение имеет величина, которая принимает конечное число $n$ значений с равными вероятностями, соответственно, вероятность каждого значения равна $\frac{1}{n}$. Отсюда уже интуитивно понятно, как будет выглядеть график плотности и гистограмма. При увеличении разности $b-a$, высота столбиков будет уменьшаться, чтобы в суммме вссе равно давать единицу. Все согласуется с теорией $\xi \sim U[a, b], P(\xi = k \in [a, b]) = \frac{1}{n}$
