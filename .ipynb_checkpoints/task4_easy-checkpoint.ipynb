### Задача 1.

В этой задаче нужно визуализировать *центральную предельную теорему*. Перед выполнением этой задачи настоятельно рекомендуется посмотреть <a href="https://mipt-stats.gitlab.io/courses/ad_fivt/LLN.html">ноутбук</a> с лекции про закон больших чисел. За использование циклов оценка может быть существенно снижена.

**a).** Пусть $\xi_1, ..., \xi_n$ &mdash; независимые случайные величины из распределения $Exp(\lambda)$. Согласно центральной предельной теореме выполнена сходимость $$Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\longrightarrow} \mathcal{N}(0, 1),$$ где $X_n = \sum\limits_{i=1}^n \xi_i$. Вам нужно убедиться в этом, сгенерировав множество наборов случайных величин и посчитав по каждому из наборов величину $Z_n$ в зависимости от размера набора. 

Сгенерируйте 500 наборов случайных величин $\xi_1^j, ..., \xi_{300}^j$ из распределения $Exp(1)$. По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора. Для этого среднего посчитайте величину $$Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}.$$
Обратите внимание, что нужно взять теоретические математическое ожидание и дисперсию, а не вычислять их по выборке.

$E X_{jn} = E \left[ \sum\limits_{i = 1}^n \xi_i^j \right] = \sum\limits_{i = 1}^n E \xi_i^j = n \cdot E \xi_1^1 = \frac{n}{\lambda}$,

$D X_{jn} = D \left[ \sum\limits_{i = 1}^n \xi_i^j \right] = \sum\limits_{i = 1}^n D \xi_i^j = n \cdot D \xi_1^1 = \frac{n}{\lambda^2}$,

$Z_{j n} = \frac{\lambda X_{j n} - n}{\sqrt{n}}$.

````python
k = 300  # Length of series in each sample
lambda_exp = 1  # Parameter of exp distribution

samples = sps.expon(0, lambda_exp).rvs(size=(500, k)).cumsum(axis=1) * lambda_exp
samples -= np.arange(1, k + 1)
samples /= np.sqrt(np.arange(1, k + 1))
````

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

````python
plt.figure(figsize=(10, 7))
for series in samples:
    plt.plot(
        np.arange(1, k + 1), 
        series, 
        alpha=0.05,
        c='red'
    )

plt.xlabel('Series size')
plt.ylabel('Transformed value')
plt.title(f'Сonvergence of Exp({lambda_exp}) in the central limit theorem')
plt.show()
````

Значения $Z_{j n}$ сходятся к некому симметричному распределению с центром в $0$.

Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{500,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$, которую тоже постройте на том же графике? Не забудьте сделать легенду.

````python
plt.figure(figsize=(10, 7))
grid = np.linspace(-3, 3, 1000)

plt.hist(
    samples[:,-1], 
    bins=10,
    alpha=0.5,
    density=True,
    label='Data'
)

plt.plot(
    grid, 
    sps.norm(0, 1).pdf(grid), 
    color='red',
    linewidth=3, 
    label='Density'
)

plt.legend()
plt.xlabel('Transformed values')
plt.ylabel('Percentage of such values')
plt.title(f'Histogram of transfomed values of Exp({lambda_exp}) on {k} step')
plt.show()
````

Гистограмма исследуемых величин крайне похожа на плотность стандартного нормального распределения, что позволяет сделать вывод о том, что для экспоненциального распредления с параметром $1$ центральная предельная теорема верна.

**b).** Выполните те же действия для распределений $U(0, 1)$ и $Pois(1)$.

Для равномерного распределения $U(0, 1)$:

$E X_{j n} = n \cdot E \xi_1^1 = \frac{n}{2}$,

$D X_{j n} = n \cdot D \xi_1^1 = \frac{n}{12}$,

$Z_{j n} = \frac{2 X_{j n} - n}{\sqrt{\frac{n}{3}}}$.

````python
samples = sps.uniform.rvs(size=(500, k)).cumsum(axis=1) * 2
samples -= np.arange(1, k + 1)
samples /= np.sqrt(np.arange(1, k + 1)) / np.sqrt(3)
````

Рассмотрим как преобразованные случайные величины изменяются по мере стремления к бесконечности.

````python
plt.figure(figsize=(10, 7))
for series in samples:
    plt.plot(
        np.arange(1, k + 1), 
        series, 
        alpha=0.05,
        c='red'
    )

plt.xlabel('Series size')
plt.ylabel('Transformed value')
plt.title('Сonvergence of U(0, 1) in the central limit theorem')
plt.show()
````

Значения $Z_{j n}$ сходятся к некому симметричному распределению с центром в $0$.

````python
plt.figure(figsize=(10, 7))
grid = np.linspace(-3, 3, 1000)

plt.hist(
    samples[:,-1], 
    bins=10,
    alpha=0.5,
    density=True,
    label='Data'
)

plt.plot(
    grid, 
    sps.norm(0, 1).pdf(grid), 
    color='red',
    linewidth=3, 
    label='Density'
)

plt.legend()
plt.xlabel('Transformed values')
plt.ylabel('Percentage of such values')
plt.title(f'Histogram of transfomed values of U(0, 1) on {k} step')
plt.show()
````

Гистограмма исследуемых величин крайне похожа на плотность стандартного нормального распределения, что позволяет сделать вывод о том, что для непрерывного равномерного распределения с параметрами $(0, 1)$ центральная предельная теорема верна.

Для равномерного распределения $Pois(1)$:

$E X_{j n} = n \cdot E \xi_1^1 = n$,

$D X_{j n} = n \cdot D \xi_1^1 = n$,

$Z_{j n} = \frac{X_{j n} - n}{\sqrt{n}}$.

````python
samples = sps.poisson(1).rvs(size=(500, k)).cumsum(axis=1).astype(float)
samples -= np.arange(1, k + 1)
samples /= np.sqrt(np.arange(1, k + 1))
````

Рассмотрим как преобразованные случайные величины изменяются по мере стремления к бесконечности.

````python
plt.figure(figsize=(10, 7))
for series in samples:
    plt.plot(
        np.arange(1, k + 1), 
        series, 
        alpha=0.05,
        c='red'
    )

plt.xlabel('Series size')
plt.ylabel('Transformed value')
plt.title('Сonvergence of Pois(1) in the central limit theorem')
plt.show()
````

Значения $Z_{j n}$ сходятся к некому симметричному распределению с центром в $0$.

````python
plt.figure(figsize=(10, 7))
grid = np.linspace(-3, 3, 1000)

plt.hist(
    samples[:,-1], 
    bins=10,
    alpha=0.5,
    density=True,
    label='Data'
)

plt.plot(
    grid, 
    sps.norm(0, 1).pdf(grid), 
    color='red',
    linewidth=3, 
    label='Density'
)

plt.legend()
plt.xlabel('Transformed values')
plt.ylabel('Percentage of such values')
plt.title(f'Histogram of transfomed values of Pois(1) on {k} step')
plt.show()
````

Гистограмма исследуемых величин крайне похожа на плотность стандартного нормального распределения, что позволяет сделать вывод о том, что для распределения Пуассона с параметром $1$ центральная предельная теорема верна.

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn import metrics
````

Подгрузим данные:

````python
data = pd.read_csv('day.csv', parse_dates=[1])
data.head()
````

Уберём признаки, которые рассматривать необходимости нет.

````python
data = data.drop(columns=['instant', 'dteday', 'yr', 'casual', 'registered'])
data.head()
````

Разделим данные на тренировочную и тестовую выборки:

````python
train, test = train_test_split(data, test_size=0.2)
````

Посмотрите по графикам, как целевой признак зависит от остальных и поймите характер зависимости целевого признака от остальных. 

````python
categorial_features = [
    'season',
    'mnth',
    'holiday',
    'weekday',
    'workingday',
    'weathersit'
]
real_features = [
    'temp',
    'atemp',
    'hum',
    'windspeed'
]
target_feature = 'cnt'

for hue in categorial_features:
    g = sns.PairGrid(train[[
        'temp',
        'atemp',
        'hum',
        'windspeed',
        'cnt',
        hue
    ]], hue=hue, diag_sharey=False, height=3)
    
    g.map_lower(sns.kdeplot, alpha=0.6)
    g.map_upper(plt.scatter, alpha=0.3)
    g.map_diag(sns.kdeplot, lw=3, alpha=0.6, 
               common_norm=False)
    
    g.add_legend()
````

Из графиков можно сделать вывод о том, что

Какие признаки категориальные, а какие &mdash; вещественные? Что нужно сделать с категориальными признаками? Выполните это. При необходимости можно сделать несложную обработку вещественных признаков.

Приведём категориальные признаки к бинарному виду и составим матрицу с чисто числовыми признаками:

````python
encoder = OneHotEncoder(drop='first', sparse=False)
train_categories = encoder.fit_transform(train[categorial_features])
X_train = np.hstack([train[real_features], train_categories])
````

Обучите линейную регрессию из `sklearn` на наших данных и посмотрите на полученные оценки коэффициентов.

````python
model = LinearRegression(fit_intercept=True)
model.fit(X_train, train[target_feature])
print(model.coef_, model.intercept_)
````

Посчитайте ошибку предсказания на тестовой выборке по метрикам RMSE, MAE, MAPE.

````python
test_categories = encoder.transform(test[categorial_features])
X_test = np.hstack([test[real_features], test_categories])
test_preds = model.predict(X_test)
````

RMSE:

````python
np.sqrt(metrics.mean_squared_error(test[target_feature], test_preds))
````

MAE:

````python
metrics.mean_absolute_error(test[target_feature], test_preds)
````

MAPE:

````python
metrics.mean_absolute_percentage_error(test[target_feature], test_preds)
````

В получившейся модели погрешность немного более тысячи, при том что сама оцениваемая величина не превышает $10000$, а MAPE так и вовсе $28\%$.

Попробуйте посчитать оценку коэффициентов с помощью формулы, полученной на занятии. В чем заключается проблема?

*Подсказка.* Подумайте о зависимостях между признаками.

На лекции была получена формула аналитического решения задачи: $\theta = (X^T X)^{-1} X^T Y$ -- вектор коэффициентов при соответствующих признаках.

````python
coefs = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ train[target_feature].to_numpy()
coefs
````

Отметим, что коэффициенты получились крайне большими, что вызывает подозрения, возможно модель слишком заточена под тренировочные данные, переобучена.

Проблема данного метода заключается в том, что если признаки линейно зависимы, то матрица $X$ вырождена, тогда вырождена матрица $X^T X$ (определитель произведения -- произведение определителей). А обратные матрицы определены только для невырожденных матриц. Значит если существуют линейно зависимые признаки, то формула просто некорректна.

Однако в реальной жизни из-за погрешностей, флуктуаций и прочих радостей лабораторных работ вещественные признаки не будут идеально линейно зависимы, а значит обратная матрица будет определена, но будет с очень большими значениями (в некотором смысле, обратной матрице к вырожденной будет матрица с бесконечным определителем).

Исправьте эту проблему, посчитайте оценки коэффициентов, а также ошибку предсказания.

Благодаря анализу графиков (и во втором пункте здравому смыслу), уберём связанные признаки:

1. temp и atemp,
2. holiday, weekday и workingday,
3. season, mnth.

````python
X_clean = data.drop(columns=[
    'holiday',
    'mnth',
    'atemp',
    'weekday',
])

train_clean, test_clean = train_test_split(X_clean, test_size=0.2)

categorial_features = [
    'season',
    'weathersit',
    'workingday',
]

real_feautures = [
    'temp',
    'hum',
    'windspeed',
]

encoder = OneHotEncoder(drop='first', sparse=False)
train_categories = encoder.fit_transform(train_clean[categorial_features])
X_clean_train = np.hstack([train_clean[[
    'temp',
    'hum',
    'windspeed',
]], train_categories])
````

Ещё раз воспользуемся предлагаемой формулой:

````python
coefs = np.linalg.inv(X_clean_train.T @ X_clean_train) @ X_clean_train.T @ train[target_feature].to_numpy()
coefs
````

Коэффициенты стали где-то на $3-4$ порядка меньше. Приведём тестирующую выбору к тому же виду, что и обучающую и вычислим предсказание.

````python
test_categories = encoder.transform(test[categorial_features])
X_test_clean = np.hstack([test_clean[[
    'temp',
    'hum',
    'windspeed',
]], test_categories])

test_pred_second = X_test_clean @ coefs
````

Вычислим метрики на тестирующей выборке:

MSE:

````python
metrics.mean_squared_error(test[target_feature], test_pred_second)
````

RMSE:

````python
np.sqrt(metrics.mean_squared_error(test[target_feature], test_pred_second))
````

MAE:

````python
metrics.mean_absolute_error(test[target_feature], test_pred_second)
````

MAPE:

````python
metrics.mean_absolute_percentage_error(test[target_feature], test_pred_second)
````

Сделайте выводы.

На занятии мы уже научились классифицировать текстовые сообщения на нормальные и спам. Теперь перед вами стоит более серьезная задача &mdash; построить систему, автоматически классифицирующую поступающие новостные сюжеты на несколько заранее заданных категорий.
 




Перед выполнением задания обязательно посмотрите <a href="https://mipt-stats.gitlab.io/courses/ad_fivt/bayes_classification.html">ноутбук</a> с занятия по классификации.

---

С помощью кода ниже загрузите встроенные в `sklearn` <a target="_blank" href="https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html#sklearn.datasets.fetch_20newsgroups">данные</a>, в которых представлены различные новостные сюжеты, разделенные на 20 тематических групп.

````python
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from collections import Counter

# удаляем заголовки и подписи
remove = ('headers', 'footers')
# зафиксируем зерно случайности
random_state = 42

# Извлекаем обучающую и тестовую части перемешивая случайным образом
data_train = fetch_20newsgroups(
    subset='train', shuffle=True, random_state=random_state, remove=remove
)
data_test = fetch_20newsgroups(
    subset='test', shuffle=True, random_state=random_state, remove=remove
)
````

Посмотрим на какое-нибудь сообщение

````python
print(data_train['data'][21])
````

Индекс тематической группы для каждого сообщения

````python
data_train['target']
````

Названия тематических групп

````python
data_train['target_names']
````

С помощью методов, изученных в рамках нашего курса, постройте классификатор новостных сюжетов на тематические группы. При построении классификатора **запрещено пользоваться тестовой выборкой**. Для настройки классификатора, например, выбора оптимальных гиперпараметров, можно разделить обучающую выборку на две части &mdash; непосредственно обучающую и валидационную. Первую часть использовать для непосредственного обучения, а вторую &mdash; для настройки, пытаясь по ней оптимизировать качество.

Данные представлены в неудобном для работы формате, приведём их к ```pandas.DataFrame```:

````python
df = pd.DataFrame(data_train.data, columns=['text'])
df['target'] = data_train.target
df.head()
````

Разделим данные:

````python
data_edu, data_validate = train_test_split(df, test_size=0.2)
````

Посмотрим на сбалансированность данных:

````python
counts = pd.value_counts(data_edu.target, sort=True)

counts.plot(
    kind='bar',
    figsize=(10, 7),
    title='Amount of articles acording to category',
    xlabel='Category number',
    ylabel='Amount'
)
plt.show()
````

Классы довольно сбалансированы, чуть меньше статей $0, 18, 19$ категорий, но максимальное отличие не превышает $40\%$, поэтому необходимости учитывать это нет.

В качестве стоп-слов выберем $100$ наиболее часто встречающихся во всех текстах.

````python
all_words = ' '.join(data_edu.text).lower().replace('.', '').replace(',', '').split()
most_common = Counter(all_words).most_common(100)
````

Заведём "мешок слов":

````python
vectorizer = CountVectorizer(
    min_df=0.04,
    max_df=0.09,
    stop_words=most_common
)
vec_data_train = vectorizer.fit_transform(data_edu.text).toarray()
vec_data_test = vectorizer.transform(data_validate.text).toarray()
````

Воспользуемся наивным байесовским классификатором `MultinomialNB`. Обучим модель на обучающей выборке и оценим точность предсказания на валидационной:

````python
multinomial_nb = MultinomialNB()
multinomial_nb.fit(vec_data_train, data_edu.text)
predictions = multinomial_nb.predict(vec_data_test)
print(f'Точность: {accuracy_score(data_validate.text, predictions) :.3}')
````

Посчитайте качество на тестовой выборке.

````python
df_test = pd.DataFrame(data_test.data, columns=['text'])
df_test['target'] = data_test.target
vec_data_test = vectorizer.transform(df_test.text).toarray()
predictions = multinomial_nb.predict(vec_data_test)
print(f'Точность: {accuracy_score(df_test.text, predictions) :.3}')
````
